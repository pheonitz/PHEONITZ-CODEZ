```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

Countplot

```python
df['Survived'].value_counts().plot(kind = 'bar')
```

```python
df['Pclass'].value_counts().plot(kind = 'bar')
```

```python
df['Sex'].value_counts().plot(kind='bar')
```

PIE CHART


```python
# auto pct shows the percentage
df['Survived'].value_counts().plot(kind='pie' , autopct = '%.2f')
```

EDA FOR NUMERICAL DATA

```python
# HISTOGRAM
# BINNING
import matplotlib.pyplot as plt
plt.hist(df['Age'] , bins = 5)

```

```python
# BOXPLOT
sns.boxplot(df['Age'])
```

MULTIVARIATE ANALYSIS AND BIVARIATE ANALYSIS

```python
# BETWEEN TWO COLUMNS OR MORE
# tips is a inbuilt datset of resturant in seaborn
tips = sns.load_dataset('tips')
tips.sample(5)
```

```python
flights = sns.load_dataset('flights')
flights.sample(5)
```

SCATTERPLOT(NUMERICAL-NUMERICAL)

```python
sns.scatterplot(x = tips['total_bill'] , y = tips['tip'])
```

MULTIVARIATE ANALYSIS WITH GENDER 

```python
sns.scatterplot(x = tips['total_bill'] , y = tips['tip'] , hue= tips['sex'])
# GENERALLY MALE HAS GIVEN MORE TIP
```

```python
sns.scatterplot(x = tips['total_bill'] , y = tips['tip'] , hue= tips['sex'] , style = tips['smoker'])
# MEANS THAT SMOKERS ARE KINDHEARTED
```

BAR PLOT - USED FOR NUMERICAL DATA AND CATEGORICAL DATA


```python
# Fare for p1 class was higher 
# female average is higher than males
# df is for titanic dataset and hue is hidden parameter
sns.barplot(x = df['Pclass'] , y = df['Fare'] , hue=df['Sex'])
```

BOXPLOT MULTIVARIATE ANALYSIS

```python
sns.boxplot(x = df['Sex'] , y = df['Age'] , hue = df['Survived'])
```

DISTPLOT FOR THOSE WHO SURVIVED AND WHO DIDNT


```python
sns.histplot(df[df['Survived'] == 0]['Age'])
sns.histplot(df[df['Survived'] == 1]['Age'])
```

HEATMAP(CATEGORICAL - CATEGORICAL)

```python
pd.crosstab(df['Pclass'] , df['Survived'])
```

```python
sns.heatmap(pd.crosstab(df['Pclass'] , df['Survived']))
```

PAIRPLOT - PLOTS THE SCATTERPLOT BETWEEN EACH NUMERICAL COLUMN

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```
